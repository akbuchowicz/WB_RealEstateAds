## Preprocessing

```python
def remove_html_tags(text):
   """Remove html tags from a string"""
   import re
   clean = re.compile('<.*?>')
   return re.sub(clean, '', text)
```

```python
desc.iloc[89][0]
```

```python
remove_html_tags(desc.iloc[89][0])
```

```python
desc_cleaned = desc.applymap(remove_html_tags)
```

```python
def removeAccents(input_text):
    strange='ąćęłńóśżź\n\r'
    ascii_replacements='acelnoszz  '
    translator=str.maketrans(strange,ascii_replacements)

    return input_text.translate(translator)
```

```python
desc_cleaned_norm = desc_cleaned.applymap(removeAccents)
```

```python
desc_cleaned_norm.head()
```

# Model

```python
sample_percent = 1
num_samples = int(len(desc_cleaned_norm) * sample_percent)

subsampled_data = desc_cleaned_norm.sample(n=num_samples, random_state=42)
```

```python
model_topic = BERTopic(calculate_probabilities=True)

docs_topic = subsampled_data['description'].astype(str).tolist()

topics_topic, probabilities_topic = model_topic.fit_transform(docs_topic)
```
